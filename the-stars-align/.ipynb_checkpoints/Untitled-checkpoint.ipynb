```python
import re
import matplotlib.pyplot as plt

f = open("coordinates.txt", "r")
points=np.zeros((324,2))

ind=0

for i in range(20):
    for point in f.readlines():
        matches = re.findall(r'<(.*?)>',point)
        coordinate=matches[0].split(',')
        velocity=matches[1].split(',')

        x=int(coordinate[0])
        y=int(coordinate[1])
        x_velo=int(velocity[0])
        y_velo=int(velocity[1])

        x+=i*100000*x_velo
        y+=i*100000*y_velo

        points[ind,0]=x
        points[ind,1]=y

        ind+=1
    
    fig = plt.figure(figsize=(15, 10))
    plt.scatter(points[:,0],points[:,1],s=1)

    plt.show;
```
